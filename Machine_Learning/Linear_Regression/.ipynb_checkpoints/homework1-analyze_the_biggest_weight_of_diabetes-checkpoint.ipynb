```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
```

### data analyze and split

```python
datasets = load_diabetes()
print(datasets.DESCR)
```

```python
target = datasets['target']
data =datasets['data']
display(data,data.shape,target,target.shape)
```

```python
# we found that there are 10 character ,we need to paint each of then
```

```python
#the first character
data1 = data[:,0]
data2 = data[:,1]
data3 = data[:,2]
data4 = data[:,3]
data5 = data[:,4]
data6 = data[:,5]
data7 = data[:,6]
data8 = data[:,7]
data9 = data[:,8]
data10 = data[:,9]
display(data1,data1.shape)
```

     - age     age in years
      - sex
      - bmi     body mass index
      - bp      average blood pressure
      - s1      tc, total serum cholesterol
      - s2      ldl, low-density lipoproteins
      - s3      hdl, high-density lipoproteins
      - s4      tch, total cholesterol / HDL
      - s5      ltg, possibly log of serum triglycerides level
      - s6      glu, blood sugar level


```python
plt.scatter(data1, target)#age in years
data1 = data1.reshape(-1,1)
model = LinearRegression()
model.fit(data1,target)
display(model.coef_,model.intercept_)
```
